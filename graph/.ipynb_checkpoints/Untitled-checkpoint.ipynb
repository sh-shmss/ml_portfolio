```python
from nltk.data import find
import gensim
word2vec_sample = str(find('models/word2vec_sample/pruned.word2vec.txt'))
model = gensim.models.KeyedVectors.load_word2vec_format(word2vec_sample, binary=False)
```

```python
import networkx as nx
import numpy as np
def find_similar(model,word,n=5):
    words = model.most_similar(positive=[word],topn=n)
    G = nx.star_graph(0)
    G.clear()
    for i,j in words:
        G.add_node(i)
        G.add_weighted_edges_from([(word, i, np.exp(j)**2)])
    for v in G.nodes():
        G.nodes[v]['word']=str(v)
    return G
    
G = find_similar(model,'university', 20)

# G=Graph.copy()
# for i in Graph.nodes():
#    G_new= find_similar(model,i,10)
#    G.add_node(G_new)
#    G.add_edge(G_new, i)
```

```python
import numpy as np
def find_similar_words(model,word,n=5):
    results = model.most_similar(positive=[word],topn=n)
    return [words for words,_ in results]

def find_similar_weights(model,word,n=5):
    results = model.most_similar(positive=[word],topn=n)
    return [np.exp(weights)**2 for _,weights in results]
```

```python
G_words = find_similar_words(model,'university', 20)
G_weights = find_similar_weights(model,'university', 20)
G_words.insert(0,'university')
G_weights.insert(0,0)
```
